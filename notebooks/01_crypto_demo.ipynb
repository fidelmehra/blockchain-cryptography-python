## 1. Hashing Primitives

```python
msg = b'Blockchain Cryptography in Python - Fidel Mehra'
print('SHA-256  :', HashEngine.sha256(msg))
print('SHA-3-256:', HashEngine.sha3_256(msg))
print('Keccak256:', HashEngine.keccak256(msg))
print('RIPEMD160:', HashEngine.ripemd160(msg))
print('BLAKE2b  :', HashEngine.blake2b(msg))
```

## 2. ECDSA Key Generation & Signing

```python
kp = ECDSAKeyPair.generate()
signer = ECDSASigner(kp)
message = b'Transfer 1 BTC to Alice'
sig = signer.sign(message)
valid = signer.verify(message, sig)
print('Public key :', kp.public_key_hex[:40], '...')
print('Signature  :', sig[:40], '...')
print('Valid      :', valid)
```

## 3. Merkle Tree

```python
leaves = [b'tx1', b'tx2', b'tx3', b'tx4', b'tx5']
tree = MerkleTree(leaves)
print('Merkle Root:', tree.root)
proof = tree.get_proof(2)
print('Proof for index 2:', proof)
print('Proof valid:', tree.verify_proof(leaves[2], proof, tree.root))
```

## 4. Block Mining (Proof-of-Work)

```python
import time
block = Block(index=1, transactions=[], previous_hash='0'*64, difficulty=3)
t0 = time.time()
block.mine()
print(f'Block mined in {time.time()-t0:.3f}s')
print('Nonce :', block.nonce)
print('Hash  :', block.hash[:40], '...')
```

## 5. Full Blockchain

```python
chain = Blockchain(difficulty=2)
# Create a UTXO transaction
txout = TransactionOutput(recipient_address=kp.address, amount=50.0)
tx = Transaction(inputs=[], outputs=[txout])
chain.add_transaction(tx)
chain.mine_pending_transactions(miner_address=kp.address)
print('Chain length:', len(chain.chain))
print('Chain valid :', chain.is_valid())
print('Balance     :', chain.get_balance(kp.address))
```
